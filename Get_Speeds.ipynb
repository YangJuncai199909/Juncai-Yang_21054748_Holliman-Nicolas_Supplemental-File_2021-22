```python
# Network Speed Test Code
import psutil
import time
import re, sys

# cited by: https://blog.csdn.net/kermit32/article/details/117393935
def processinfo(x):
    p = psutil.process_iter()
    tlp = 0
    try:
        for r in p:
            aa = str(r)
            f = re.compile(x, re.I)
            if f.search(aa):
                tlp = int(aa.split('pid=')[1].split(',')[0])
                # Retrieve a list of pids and get the pid of the passed in value
                return tlp
    except (psutil.NoSuchProcess):
            print('Ransomware process is caught, but the process does '
                  'not exist (PID: %d)' % aa.pid)

# cited by: https://blog.csdn.net/kermit32/article/details/117393935
def getNet():
    sent_before = psutil.net_io_counters().bytes_sent  # Traffic sent
    recv_before = psutil.net_io_counters().bytes_recv  # Traffic received
    time.sleep(1)
    sent_now = psutil.net_io_counters().bytes_sent
    recv_now = psutil.net_io_counters().bytes_recv
    sent = (sent_now - sent_before)/1024  # Calculate the difference after 1 second
    recv = (recv_now - recv_before)/1024
    print("Upload Rate：{0}KB/s".format("%.2f"%sent),"Download Rate：{0}KB/s".format("%.2f"%recv))

# cited by: https://blog.csdn.net/kermit32/article/details/117393935
def getinfo(tlp):
    p = psutil.Process(tlp)
    cpu_list = []
    for i in range(10):
        p_cpu = p.cpu_percent(interval=0.1)
        cpu_list.append(p_cpu)
        cpu = 0.00
        cpu = float(sum(cpu_list))/len(cpu_list)/10
        # Loop through the cpu usage values 10 times and take the average

    try:
        pid = p.pid
        name = p.name()
        Memory = p.memory_percent(memtype="rss") / 2
        localtime = time.strftime('%H:%M:%S', time.localtime(time.time()))
        status = p.status()
        # Get process pid ,process name ,process memory
    except IOError as e:
        print(e)
    else:
        # return pid, name, Memory, cpu, time
        print("Time:%s" % (localtime), "PID:%s" % (pid), "Name:%s" % (name),
              "Memory=%.3f%%" % (Memory), "CPU=%.2f%%" % (cpu * 2), "Status:%s" % (status))



if __name__ == "__main__":
    while 0 < 1:
        # If the software used has changed, it needs to be reset
        # s = processinfo('wemeetapp.exe')
        s = processinfo('Teams.exe')
        getinfo(s)
        getNet()
        print('-' * 32)
        if False:
            print("Opening the program")
        else:
            continue
```
